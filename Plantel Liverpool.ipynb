```python
from tkinter import *

######## Funcionalidades do sistema #############
import pyodbc

dados_db = ("Driver={SQLite3 ODBC Driver};Server=localhost;Database=jogadores.db")
conexao_db = pyodbc.connect(dados_db)
cursor = conexao_db.cursor()

def adicionar_jogador():
    if len(nome_jogador.get()) == 0:
        # exibir uma mensagem -> nome jogador inválido
        
        # deletar tudo da caixa de texto
        caixa_texto.delete("1.0", END)
        
        # escrever na caixa de texto
        caixa_texto.insert("1.0", f"Não é possível adicionar o jogador!")
        
        # finalizar a função
        return
    
    # Adicionando informações no Banco de Dados:

    cursor.execute(f'''
    INSERT INTO jogadores (Nome, Idade, Nacionalidade, Posicao)
    VALUES
    ('{nome_jogador.get()}', {idade_jogador.get()}, '{nacionalidade_jogador.get()}', '{posicao_jogador.get()}')
    ''')
    cursor.commit()
    
    # deletar tudo da caixa de texto
    caixa_texto.delete("1.0", END)
    
    cursor.commit()
    
    # escrever na caixa de texto
    caixa_texto.insert("1.0", f"{posicao_jogador.get()} {nome_jogador.get()} adicionado com sucesso!")
    
def deletar_jogador():
    
    #print("deletar_jogador")
    
    if len(nome_jogador.get()) == 0:
        # exibir uma mensagem -> nome jogador inválido
        
        # deletar tudo da caixa de texto
        caixa_texto.delete("1.0", END)
        
        # escrever na caixa de texto
        caixa_texto.insert("1.0", f"Não é possível deletar o jogador!")
        
        # finalizar a função
        return
    
    # Deletar o jogador
    cursor.execute(f'''
    DELETE FROM jogadores 
    WHERE Nome="{nome_jogador.get()}"
    ''')
    cursor.commit()
    
    # Limpar/deletar tudo da caixa de texto
    caixa_texto.delete("1.0", END)
    
    # Escrever na caixa de texto
    caixa_texto.insert("1.0", f"{nome_jogador.get()} deletado com sucesso!") 
    

def procurar_jogador():
    if len(nome_jogador.get()) == 0:
        # exibir uma mensagem -> nome jogador inválido
        
        # deletar tudo da caixa de texto
        caixa_texto.delete("1.0", END)
        
        # escrever na caixa de texto
        caixa_texto.insert("1.0", "Não é possível visualizar as informações do jogador!")
        
        # finalizar a função 
        return
    
    # Pesquisar pelo jogador
    
    cursor.execute(f'''
    SELECT * FROM jogadores WHERE Nome="{nome_jogador.get()}"
    ''')
    valores = cursor.fetchall()
    
    texto = ""
    for nome, idade, nacionalidade, posicao in valores:
        texto = texto + f'''----------------------------------
Nome: {nome}
Idade: {idade}
Nacionalidade: {nacionalidade}
Posição: {posicao}  
---------------------------------
'''

    # deletar tudo da caixa de texto
    caixa_texto.delete("1.0", END)
    # escrever na caixa de texto
    caixa_texto.insert("1.0", texto)

def listar_jogadores():    
    # deletar tudo da caixa de texto
    caixa_texto.delete("1.0", END)

    # Listar todos os jogadores:
    
    cursor.execute(f'''
    SELECT * FROM jogadores Order by Nome ASC''')
    
    valores = cursor.fetchall()
    
    texto = ""
    for nome, idade, nacionalidade, posicao in valores:
        texto = texto + f'''----------------------------------
Nome do jogador: {nome}
Idade: {idade}
Nacionalidade: {nacionalidade}
Posição: {posicao}
----------------------------------
\n
'''

    # deletar tudo da caixa de texto
    caixa_texto.delete("1.0", END)
    
    # escrever na caixa de texto
    caixa_texto.insert("1.0", texto)

######### Criação da Janela ##################
    
window = Tk()
window.title('Plantel Liverpool 2023')

window.geometry("855x590")
window.configure(bg = "#ffffff")
canvas = Canvas(
    window,
    bg = "#ffffff",
    height = 590,
    width = 855,
    bd = 0,
    highlightthickness = 0,
    relief = "ridge")
canvas.place(x = 0, y = 0)

background_img = PhotoImage(file = f"imagens/background.png")
background = canvas.create_image(
    0, 0, anchor='nw',
    image=background_img)

# entry0_img = PhotoImage(file = f"imagens/img_textBox0.png")
# entry0_bg = canvas.create_image(
#     308.0, 342.0,
#     image = entry0_img)

caixa_texto = Text(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

caixa_texto.place(
    x = 479, y = 456,
    width = 274,
    height = 110)

# entry1_img = PhotoImage(file = f"imagens/img_textBox1.png")
# entry1_bg = canvas.create_image(
#     357.0, 235.0,
#     image = entry1_img)

idade_jogador = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

idade_jogador.place(
    x = 580, y = 301, # Idade
    width = 170,
    height = 26)

# entry2_img = PhotoImage(file = f"imagens/img_textBox2.png")
# entry2_bg = canvas.create_image(
#     357.0, 190.0,
#     image = entry2_img)

nacionalidade_jogador = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

nacionalidade_jogador.place( # Nacionalidade
    x = 580, y = 346,
    width = 170,
    height = 26)

# entry3_img = PhotoImage(file = f"imagens/img_textBox3.png")
# entry3_bg = canvas.create_image(
#     357.0, 145.0,
#     image = entry3_img)

posicao_jogador = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

posicao_jogador.place( # Posição do jogador
    x = 580, y = 391,
    width = 170,
    height = 26)

# entry4_img = PhotoImage(file = f"imagens/img_textBox4.png")
# entry4_bg = canvas.create_image(
#     357.0, 102.0,
#     image = entry4_img)

nome_jogador = Entry(
    bd = 0,
    bg = "#ffffff",
    highlightthickness = 0)

nome_jogador.place( # Nome do jogador
    x = 580, y = 258,
    width = 170,
    height = 26)

img0 = PhotoImage(file = f"imagens/img0.png")
b0 = Button(
    image = img0,
    borderwidth = 0,
    highlightthickness = 0,
    command = deletar_jogador,
    relief = "flat")

b0.place(
    x = 479, y = 176, # Deletar jogador
    width = 123,
    height = 30)

img1 = PhotoImage(file = f"imagens/img1.png")
b1 = Button(
    image = img1,
    borderwidth = 0,
    highlightthickness = 0,
    command = listar_jogadores,
    relief = "flat")

b1.place(
    x = 633, y = 116, # Listar jogador
    width = 121,
    height = 30)

img2 = PhotoImage(file = f"imagens/img2.png")
b2 = Button(
    image = img2,
    borderwidth = 0,
    highlightthickness = 0,
    command = procurar_jogador,
    relief = "flat")

b2.place(
    x = 633, y = 177, # Procurar jogador
    width = 121,
    height = 30)

img3 = PhotoImage(file = f"imagens/img3.png")
b3 = Button(
    image = img3,
    borderwidth = 0,
    highlightthickness = 0,
    command = adicionar_jogador,
    relief = "flat")

b3.place(
    x = 479, y = 116, # Adicionar jogador
    width = 121,
    height = 30)

window.resizable(False, False)
window.mainloop() # loop infinito que exibe a tela pra você, só é finalizado quando fechamos a janela...

cursor.close()
conexao_db.close()
```
